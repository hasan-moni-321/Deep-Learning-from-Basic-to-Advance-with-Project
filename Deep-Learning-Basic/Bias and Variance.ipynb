# Overfitting and underfitting

overfitting:-
    overfitting means when our algorithm learn too much and provide a very good accuracy in training data but when in test data algorithm provide a very poor accuracy.
    
Underfitting:-
    underfitting means when our algorithm learn too poor both in training and test data and provide a poor accuracy for both training and test data
    

# Bias and Variance for Regression

Bias:-                                                                                                                 
    bias means the error in training data in short.
    
Variance:-                                                                                                                     
    variance means the error in test data in short.

It is clear that an underfitted model have high bias and high variance because both training and test accuracy is less

In good model have low bias and low variance because of good fitting in training and test data

In ovefitted model have low bias and high variance because model learn properly from training data but not give a good prediction in test data
