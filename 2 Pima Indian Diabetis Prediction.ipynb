# Predicting Diabetis 

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd

seed = 7
numpy.random.seed(seed)
```

##### Loding Dataset

```python
dataset = pd.read_csv("/home/hasan/DATA SET/pima-indians-diabetes-database/diabetes.csv", delimiter=",")
```

```python
dataset.head()
```

##### Split Dataset

```python
X = dataset.iloc[:,0:8]
Y = dataset.iloc[:,8]
```

##### Model

```python
model = Sequential()
model.add(Dense(50, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(25, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
#summary of the model
model.summary()
```

```python
#compile the Model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

##### Fit the Model

```python
model.fit(X, Y, nb_epoch=150, batch_size=10)
```

##### Evaluate the Model

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
